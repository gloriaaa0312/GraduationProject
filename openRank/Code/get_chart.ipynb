```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('../Result/all_skl_results.csv')
print(data)
```

```python
metrics = data['Metric'].unique()
models = data['Model'].unique()

pivoted_data = data.pivot_table(index='Model', columns='Metric', values='Value')
# print(pivoted_data['MSE'])
for i in metrics:
    print(i)
    pivoted_data[i].plot(kind='bar')
    plt.show()
```
